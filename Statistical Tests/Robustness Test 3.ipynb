```python
import pandas as pd
import numpy as np
```

```python
NBEATSx=pd.read_excel("NBEATSx forecasts.xlsx", index_col="Unnamed: 0")
```

```python
NHITS=pd.read_excel("NHITS forecasts.xlsx", index_col="Unnamed: 0")
```

```python
TFT=pd.read_excel("TFT forecasts.xlsx", index_col="Unnamed: 0")
```

```python
LSTM=pd.read_excel("LSTM forecasts.xlsx", index_col="Unnamed: 0")
```

# Results

```python
models = [NHITS,NBEATSx,TFT, LSTM]
models_names = ["NHITS","NBEATSx","TFT", "LSTM"]
for j in range(len(models)):
  print(f"{models_names[j]}")
  for i in range(len(NBEATSx.columns)):
    print(f"Mean of {models[j].columns[i]}: {round(100*models[j][models[j].columns[i]].mean(),3)}%")
    print(f"Standard Deviation of {models[j].columns[i]}: {round(100*models[j][models[j].columns[i]].std(),3)}%")
```

# T-tests

```python
from scipy import stats
models = [NBEATSx,TFT, LSTM]
models_names = ["NBEATSx","TFT", "LSTM"]
for j in range(len(models)):
  print(f"{models_names[j]}")
  for i in range(len(NBEATSx.columns)):
    print(f"{models[j].columns[i]}: {stats.ttest_ind(models[j][models[j].columns[i]], NHITS[models[j].columns[i]])}")
```

# F-tests

```python
import numpy as np
import scipy

def f_test(x, y):
    x = np.array(x)
    y = np.array(y)
    f = np.var(x, ddof=1)/np.var(y, ddof=1) #calculate F test statistic
    dfn = x.size-1 #define degrees of freedom numerator
    dfd = y.size-1 #define degrees of freedom denominator
    p = 1-scipy.stats.f.cdf(f, dfn, dfd) #find p-value of F test statistic
    return f, p

models = [NBEATSx,TFT, LSTM]
models_names = ["NBEATSx","TFT", "LSTM"]
for j in range(len(models)):
  print(f"{models_names[j]}")
  for i in range(len(NBEATSx.columns)):
    print(f"{models[j].columns[i]}: {f_test(models[j][models[j].columns[i]], NHITS[models[j].columns[i]])}")
```
